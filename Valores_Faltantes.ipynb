```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive

# Cargar dataset Titanic
drive.mount('/content/drive')
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/train.csv")
```

```python
# --- Conteo de valores nulos ---
faltantes = df.isnull().sum()
faltantes = faltantes[faltantes > 0].sort_values(ascending=False)

print("=== Valores faltantes por variable ===")
print(faltantes)
print("\nPorcentaje de valores faltantes:")
print((faltantes / len(df) * 100).round(2))
```

```python
# --- Heatmap de valores faltantes ---
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de valores faltantes en Titanic")
plt.show()
```

```python
# --- Barplot con % de faltantes ---
faltantes_pct = (faltantes / len(df) * 100).round(2)
plt.figure(figsize=(8,5))
sns.barplot(x=faltantes_pct.index, y=faltantes_pct.values, palette="mako")
plt.ylabel("% de valores faltantes")
plt.title("Porcentaje de valores faltantes por variable")
plt.show()
```
