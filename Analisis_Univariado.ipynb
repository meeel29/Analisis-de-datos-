```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# --- Configuración de estilo de gráficos --- 
sns.set(context="notebook", style="whitegrid")

df = pd.read_csv("./train.csv")

# Desde Google Colab, montar desde Google Drive
# from google.colab import drive
#drive.mount('/content/drive')
#df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/train.csv")
```

### Variables númericas

#### Age

```python
# --- Resumen estadístico de la variable Age ---
df['Age'].describe().round(2)
```

```python
# --- Distribución de la variable Age ---
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], kde=True, bins=30, color="teal")
plt.title("Distribución de Age")
plt.xlabel("Edad"); plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
```

```python
# --- Boxplot de Age --- 
plt.figure(figsize=(5,3))
sns.boxplot(x=df['Age'], color="salmon")
plt.title("Boxplot de Age")
plt.xlabel("Edad")
plt.tight_layout()
plt.show()
```

```python
# --- Asimetría y curtosis de Age ---
s = df['Age'].dropna()
print("Skewness:", round(skew(s), 3))
print("Kurtosis:", round(kurtosis(s), 3))
```
