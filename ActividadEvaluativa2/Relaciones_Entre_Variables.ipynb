```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Configuración de estilo de gráficos --- 
sns.set(context="notebook", style="whitegrid")

df = pd.read_csv("./train.csv")

# Desde Google Colab, montar desde Google Drive
# from google.colab import drive
#drive.mount('/content/drive')
#df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/train.csv")
```

```python
# --- Pairplot de variables numéricas coloreado por supervivencia ---
num_vars = ['Age','Fare','SibSp','Parch','Pclass','Survived']

sns.pairplot(df[num_vars], hue="Survived", diag_kind="kde", palette="husl")
plt.suptitle("Relaciones entre variables numéricas y Supervivencia", y=1.02)
plt.show()
```
